* days_exposition - 20518 строк (3181-13.0% Пропущенных значений) **float64->int требует форматирования данных и заполнения пропущенных значений, т.к. 13% для удаления это очень много. Просто заполнить медианой по всем датафрейму будет скорее всего не верно, т.к. в некоторых районах квартиры имеют больший спрос и продаются быстрее чем где то в глуши. Необходимо сгруппировать данные по районам и тогда заполнить медианой по каждому населенному пункту.

```python
data['days_exposition'].isna().sum()
```

```python
# Узнаем в каких населенных пунктах есть значения NaN
print('Пропущенные значенмя по населенным пунктам СПб')
data.groupby(data.loc[data['days_exposition'].isna()==True,'locality_name'])['locality_name'].count(). \
                                                                                 sort_values(ascending=False).head(10)
```

```python
# Заполним пропуски медианой
print('Median days_exposition по населенным пунктам: ')
median_columns = data.groupby('locality_name')['days_exposition'].median().sort_values()
median_columns = median_columns.dropna()
display(median_columns)

# Функция для присваивания пропуску значения медианы
def loc_med_days(loc_name):
    if loc_name not in median_columns:
        return np.nan
    return median_columns[loc_name]
```

```python
# Заменяем nan на среднее 
data.loc[data['days_exposition'].isna()==True, 'days_exposition'] = data['locality_name'].apply(loc_med_days)
```

```python
# Проверим результат
data['days_exposition'].isna().sum()
```

Прпуски еще остались. Чтобы поменять тип столбца на int необходимо окончательно избавиться от них. Можем поставить 0, но это может оказать влияние на исследование. Более логично будет присвоить среднее значение столбца, так значеня не должны оказать влияния на результат.

```python
# заполним пропуски медианой всего столбца
med_day = data['days_exposition'].median()
data['days_exposition'] = data['days_exposition'].fillna(med_day)
```

```python
data['days_exposition'].isna().sum()
```

Попуски в living_area заполним, по среднему по категориям в соотношении жилой площади к общей.

```python
# Создадим функции с категориями по столбцам total_area и rooms
def category_area(area):
    if area <= 40:
        return 'маленкая'
    elif 40 < area <=69 :
        return 'средняя'
    else:
        return 'большая'
data['category_area'] = data['total_area'].apply(category_area)
```

```python
def category_rooms(rooms):
    if rooms <= 1:
        return 'маленькая'
    elif 1 < rooms <= 3 :
        return 'средняя'
    else:
        return 'большая'
data['category_rooms'] = data['rooms'].apply(category_rooms)
```

```python
# Создадим сводную табицу для заполнения медианой
area_pivot_table = data.pivot_table(index = ['category_rooms', 'category_area'],
                                       columns = 'studio',
                                       values = 'living_area', aggfunc = 'median' ).dropna(axis = 0)
```

```python
# Создадим функцию для заполнения пропусков медианой
def columns_fillna(row, pivot_table,aim_column, index_columns):
    if pd.isnull(row[aim_column]):
        index = (index_columns[0],index_columns[1])
        if index in pivot_table.index:
            return pivot_table.loc[index][0]
        else:
            return data[aim_column].median()
    else:
        return row[aim_column]
```

#### Столбец "cityCenters_nearest" (расстояние до центра города (м)) 

* cityCenters_nearest - 18180 (5519-23.0% Пропущенных значений) float64->int. Они могут отустствовать по причине того, что относительно конкретно взятого населенного пункта/точки искались расстояние в определенном радиусе. Если в пределах этого радиуса ни аэропорта, ни парков не было, то и указать в этом столбце было нечего. Заменим медианой по нас.пункту

```python
print('Количество NaN в столбце cityCenters_nearest:', data['cityCenters_nearest'].isna().sum())
```

```python
# Создадим сводную таблицу с медианой по локации квартиры
cityCenters_pivot = data.pivot_table(index=['locality_name'], values=['cityCenters_nearest'], aggfunc='median')
dict_space = {}
data['cityCenters_nearest'].fillna(0,inplace=True)
mediana_locality = data['cityCenters_nearest'].median()
cityCenters_pivot.head()
```

```python
# создадим словарь сопоставлений по локации
for i in data['locality_name'].unique():
    if i in cityCenters_pivot.index:
        dict_space[i] = cityCenters_pivot.loc[i,'cityCenters_nearest']
    else:
        dict_space[i] = mediana_locality

# напишем функцию
def func(row):
    if row.cityCenters_nearest == 0:
        row.cityCenters_nearest = dict_space[row.locality_name]
        return row.cityCenters_nearest
    else:
        return row.cityCenters_nearest
```

```python
# выполним замену значений 
data.cityCenters_nearest = data.apply(func, axis = 1)  
```

```python
print('Количество NaN в столбце cityCenters_nearest:', data['cityCenters_nearest'].isna().sum())
```

Ну и наконец для красивого и более понятного дальнейшего анализа сгруппируем населеные пункты по удаленности от центра. За центр возьмем Спб.

Сгруппируем по следующим признакам:

* 0-2 км от центра (категория 'самый центр');
* 2-5 км от центра (категория 'центр');
* 5-17 км от центра (категория 'близко');
* 17-32 км от центра (категория 'в пределах КАД ');
* более 32 км от центра (категория 'область ')

```python
def grouping_cityCenters_nearest(row):
    if row['cityCenters_nearest'] < 2000:
        return 'самый центр'
    elif row['cityCenters_nearest'] < 5000:
        return 'центр'
    elif row['cityCenters_nearest'] < 17000:
        return 'в пределах города'
    elif row['cityCenters_nearest'] < 32000:
        return 'в пределах КАД'
    elif row['cityCenters_nearest'] >= 32000:
        return 'область'
    
data['distance_from_center_km'] = data.apply(grouping_cityCenters_nearest, axis=1)
```

#### * Поиск редких и выбивающихся значений

Уберите редкие и выбивающиеся значения. Опишите, какие особенности обнаружили.

```python
data.describe()
```

Удалять будем основываясь на квартили

```python
def anomaly_removal(column,data):
    q25=np.array(data[column].describe(percentiles=[.25,.75]).loc['25%'])
    q75=np.array(data[column].describe(percentiles=[.25,.75]).loc['75%'])
    first_part=q25-1.5*(q75-q25)
    second_part=q75+1.5*(q75-q25)
    count = 0
    leng = 1
    while leng > 0:
        index_del= []
        for index, element in enumerate(data[column]):
            if first_part > element or second_part< element:
                index_del.extend(data[data[column] == element].index)

        index_del= set(index_del)
        leng = len(index_del)
        count += 1
        print('Количество строк, выбранных для удаления: {}. Количество итераций {}'.format(leng,count))
        print(leng)
        data.drop(index_del,axis=0,inplace=True)
```

```python
anomaly_removal('last_price',data)
```

```python
anomaly_removal('ceiling_height',data)
```

```python
anomaly_removal('days_exposition',data)
```

Коэффициент корреляции Пирсона измеряет линейную связь между переменными. Его значение можно интерпретировать следующим образом:

* +1 – Полная положительная корреляция
* +0.8 – Сильная положительная корреляция
* +0.6 – Умеренная положительная корреляция
* 0 – никакой корреляции вообще
* -0,6 – Умеренная отрицательная корреляция
* -0,8 – Сильная отрицательная корреляция
* -1 – Полная отрицательная корреляция

   Можем сказать, что сильнее всего цена на квартиру зависит от цены за кв метр (ну тут к бабке не ходи и ежу понятно), от общей площади квартиры (что в принципе очевидно, так как стоимость обычно зависит от цены за кв метр, соответсвенно чем больше метров тем выше стоимость), от площади жилой площади и площади кухни (ну тут не будем повторятся см предыдущий пункт), а также от количества комнат (это в принципе тоже ожидаемо, т.к. обычно чем больше общ площадь тем больше комнат, бывают конечно варианты на 100кв метров со свободной планировкой, вот там изначально одна комната), так же влияет на стоимость высота потолков. Небольшое влияние оказывает присутствие парков/водоемов рядом (тоже понятно, больше желание жить рядом с парковой зоной нежели с автострадой).

  Также влияние оказывает удаленность от центра - чем дальше от центра, тем ниже цена. Ожидаемо. Таким же образом небольшое влияние оказывает удаленность от аэропорта.
